Verificar os tipos de dados;


```python
dt_p1.info()
```

Identificar colunas numéricas;


```python
dt_p1['Valor da compra'] = dt_p1['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','').strip())
```

```python
dt_p1['Valor da compra'] = dt_p1['Valor da compra'].astype(np.float64)
```

textos

```python
dt_p1['Cliente'] = dt_p1['Cliente'].str.lower()
dt_p1['Cliente'].head()
```

```python
dt_p1['Cliente'] = dt_p1['Cliente'].str.replace('[^a-zA-Z\']',' ', regex= True)
dt_p1['Cliente']
```

```python
dt_p1
```

```python
dt_p1['Data de venda']= pd.to_datetime(dt_p1['Data de venda'],format='%Y/%d/%m')
```

```python
dt_p1
```

```python
vt = dt_p1.groupby(dt_p1['Cliente'])['Valor da compra'].sum()
```

O objetivo desse projeto é realizar uma análise dos resultados de um evento com os clientes de uma empresa de vendas online. Foi coletado um conjunto de dados que contém os clientes que mais gastaram com produtos dentro de 5 dias de vendas, que é o período de duração do evento. Essa análise vai identificar o cliente com a maior compra na semana, que irá receber um prêmio da loja, e posteriormente, ela pode ajudar a empresa a criar novas estratégias para atrair mais clientes.

```python
vt.max()
```

```python
vt.idxmax()
```

```python
cliente_maior_compra = vt.idxmax()
valor_maximo_compra = vt.max()

print(f"O cliente com a maior compra é: {cliente_maior_compra}")
print(f"O valor da maior compra é: R$ {valor_maximo_compra:,.2f}")
```

```python
import pandas as pd
```

```python
dt_p2 = pd.read_json('/content/dados_locacao_imoveis.json')
```

```python
dt_p2 = pd.json_normalize(dt_p2['dados_locacao'])
```

Remover os dados em listas dentro do DataFrame;

```python
dt_p2.info()
```

```python
dt_p2
```

```python
dt_p2=  dt_p2.explode(col_p2[1:])
```

```python
dt_p2
```

Identificar colunas numéricas;


```python
dt_p2['valor_aluguel'] = dt_p2['valor_aluguel'].apply(lambda x: x.replace("$",'').replace("reais",'').replace(",",'').strip())
```

```python
for index, row in dt_p2.iterrows():
    lengths = {col: len(row[col]) for col in ['datas_combinadas_pagamento', 'datas_de_pagamento', 'valor_aluguel']}
    if len(set(lengths.values())) > 1: # If lengths are not all the same
        print(f"Row {index} has inconsistent list lengths: {lengths}")
```

```python
dt_p2['valor_aluguel'] = dt_p2['valor_aluguel'].astype(np.float64)
```

```python
dt_p2
```

```python
dt_p2
```

```python
dt_p2['apartamento']=dt_p2['apartamento'].apply(lambda x : x.replace('(blocoAP)',''))
dt_p2['apartamento']
```

```python
dt_p2['apartamento'] = dt_p2['apartamento'].str.replace('\(BLOCO AP\)','',regex=True)
```

```python
dt_p2['apartamento']
```

```python
dt_p2
```

```python
dt_p2['datas_combinadas_pagamento']= pd.to_datetime(dt_p2['datas_combinadas_pagamento'], format='%d/%m/%Y')
```

```python
dt_p2['datas_de_pagamento']= pd.to_datetime(dt_p2['datas_de_pagamento'],format = '%d/%m/%Y')
```

```python
dt_p2['atraso'] = (dt_p2['datas_de_pagamento'] - dt_p2['datas_combinadas_pagamento']).dt.days
```

```python
dt_p2
```

```python
media_atraso = dt_p2.groupby(dt_p2['apartamento'])['atraso'].mean()
```
