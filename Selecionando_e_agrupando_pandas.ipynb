Remodelando os dados wide and long. O formato Wide , sao várias colunas enquanto o formato long sao várias linhas. O que acontece agora é usar o metodo melt, transformar os anos em linha e os valores de emissao em linhas. Para cada coluna de ano tem um valor de emissao referente a cada gas, vai ser tipo o 1970 em linha e varias emissoes para cada gas em várias linhas para o mesmo ano 1970. O que estamos interressados sao os gases nao importa os anos

Além disso precisamos manter fixo as outras colunas. Para isso, precisamos selecionar as colunas no formato de lista

```python
colunas_info = list(emissoes_gases.loc[:,'Nível 1 - Setor' : 'Produto'].columns)
```

```python
colunas_emissao = list(emissoes_gases.loc[:,1970:2021])
```

```python
emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name='Ano',value_name='Emissoes')
```

```python
emissoes_ano = emissoes_gases.melt(id_vars=colunas_info, value_vars=colunas_emissao, var_name='Ano',value_name='Emissoes')
```

```python
emissoes_ano.groupby('Gás')[['Emissoes']].sum().sort_values('Emissoes')
```

```python
emissao_por_gas =emissoes_ano.groupby('Gás')[['Emissoes']].sum().sort_values('Emissoes', ascending =False)
emissao_por_gas.plot(kind= 'barh', figsize=(10,6));
```

Note que Co2 é bem superior a todos os outros. Diante disso, é necessário somar sobre todos os dados de co2 e verificar a porcentagem em relação ao resto

```python
soma_emissao_co2= emissao_por_gas.iloc[0:9].sum()
soma_emissao_outros = emissao_por_gas.sum()
porcentagem = soma_emissao_co2/soma_emissao_outros * 100
print(f'A porcentagem de co2 em relação aos outros gases é { float((porcentagem).iloc[0])} % ')
```
